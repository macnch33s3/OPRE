# 1D-Bin-Packing-Problem

Das 1D-Binpacking-Problem ist eine Optimierungsaufgabe, bei der eine Menge von Objekten mit unterschiedlichen Grössen in eine minimale Anzahl von Behältern ("Bins") fester Kapazität gepackt werden soll. Ziel ist es, die Objekte so zu verteilen, dass kein Behälter die Kapazität überschreitet und möglichst wenig Behälter benötigt werden.

#### **Aufgabe**
Ein Distributionsunternehmen in der Elektronikbranche verpackt Pakete verschiedener Größen in Versandkartons. Die Kartons haben eine feste vorgegebene Höhe. <br>
Die Pakete, die es zu versenden gilt, passen von der Grundfläche her exakt in die Versandkartons, haben aber unterschiedliche Höhen und müssen platzsparend in die Kartons verpackt werden, um die Anzahl der benötigten Kartons zu minimieren. 

<br>
*1D-Bin-Packing Lösung, die 2 Container mit Höhe 10 benötigt*

Ihnen liegen nachfolgende Problem-Instanzen vor. Sie werden von der Geschäftsleitung beauftragt, für diese wichtigen Problem-Instanzen möglichst gute und gültige Lösungen zu entwickeln. 

*Tipp*: Es gibt in jedem der Beispiele unterschiedliche Ansätze, die zum Ziel führen. Ggf. kann es sogar sinnvoll sein, eingangs mit unterschiedlichen zu experimentieren.
