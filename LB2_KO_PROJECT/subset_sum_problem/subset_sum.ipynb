# Subset Sum Problem

Das Subset-Sum-Problem ist eine spezielle Variante des Knapsack-Problems. Beschränkt man sich beim klassischen Knapsack-Problem lediglich darauf, den Rucksack möglichst voll zu packen und ignoriert die Werte der Objekte, entsteht ein sogenanntes "Subset Sum Problem".
Es befasst sich mit der Aufgabe, aus einer gegebenen Menge von Objekten eine Teilmenge auszuwählen, deren Gesamtvolumen (oder andere kumulierbare Eigenschaft) genau einem vorgegebenen Zielwert entspricht oder diesem möglichst nahe kommt. Jedes Objekt hat dabei nur ein Volumen, jedoch keinen zusätzlichen Wert.

#### **Aufgabe**

Ein Chemieunternehmen verarbeitet die gleichen Chemikalien verschiedener Lieferanten (verschiedene Fassgrösse) in einem Batchprozess, der die Substanzen homogenisiert, auf einheitliche Korngrösse bringt und thermisch vorbehandelt. Der Batchreaktor hat eine maximale Füllmenge kann aber auch teilgefüllt betrieben werden, was natürlich mehr kostet. Es sollen immer nur ganze Fässer eingefüllt werden. Ziel ist es, eine Kombination der verfügbaren Fässer auszuwählen, um den Batchreaktor so effizient wie möglich zu befüllen, sodass die Gesamtkapazität möglichst vollständig genutzt wird, ohne die Kapazität zu überschreiten. Gleichzeitig darf kein Fass angebrochen, sondern muss immer vollständig entleert werden (nur mit ganzen Fässern rechnen).
