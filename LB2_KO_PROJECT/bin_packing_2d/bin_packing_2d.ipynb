# Offline 2D-Bin-Packing-Problem

Das Offline 2D-Binpacking-Problem ist ein kombinatorisches Optimierungsproblem, das sich mit der Aufgabe befasst, eine gegebene Menge von rechteckigen Objekten so effizient wie möglich auf eine oder mehrere rechteckige Flächen (Bins) zu platzieren (Sie können sich das so vorstellen, dass alle Pakete gleich hoch sind, oder dass sie im Versandkarton nicht gestapelt werden sollen, um Beschädigungen zu vermeiden). Dabei kennen wir im Vorhinein alle Objekte und ihre Dimensionen. Ziel ist es, die Anzahl der genutzten Bins zu minimieren. Alle Informationen über die Objekte und Bins sind im Voraus bekannt, daher der Zusatz "offline".


#### **Aufgabe**

Ein Anbieter von Ersatzteilen bietet eine breite Palette von Ersatzteilen an, die in unterschiedlich grossen, rechteckigen Schachteln verfügbar sind. Für den Versand sollen die Schachteln einer Bestellung in standardisierte, grössere Versandkartons verpackt und versandt werden. Der Anbieter möchte die Anzahl der verwendeten Versandkartons minimieren, um die Versandkosten möglichst tief zu halten (weitere Infos siehe folgender Beitrag: https://lamarr-institute.org/de/blog/bin-packing/).
