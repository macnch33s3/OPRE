# Traveling Salesman Problem

Das Traveling Salesman Problem (TSP) ist ein klassisches Optimierungsproblem, bei dem ein Verkäufer eine Reihe von Städten besuchen muss, wobei jede Stadt genau einmal besucht werden soll. Ziel ist es, die kürzeste mögliche Route zu finden, die alle Städte umfasst und wieder zum Startpunkt zurückkehrt. Die Richtung (Drehsinn) zum Abfahren der Route spielt dabei keine Rolle. 

#### **Aufgabe**

Ein Hersteller von Elektronikkomponenten verwendet in der Produktion von Leiterplatten einen automatisierten Bohrroboter, um Löcher für elektrische Verbindungen und Komponenten auf der Platine zu bohren. Auf jeder Platine befinden sich mehrere Bohrpunkte, an denen der Roboter Löcher präzise setzen muss. Da die Effizienz des Fertigungsprozesses eine wichtige Rolle spielt, soll die Gesamtlänge des Pfades, den der Roboter beim Abfahren der Bohrpunkte zurücklegt, minimiert werden. Zur Berechnung der Entfernungen ist die Euklidische Distanz massgebend.

Ihnen liegen nachfolgende Problem-Instanzen vor. Sie werden von der Geschäftsleitung beauftragt, für diese wichtigen Problem-Instanzen möglichst gute und gültige Lösungen zu entwickeln.
