Folgendes Skript zeigt, wie mit ipywidgets in Jupyter Notebooks kleine Apps entwickelt werden k철nnen. Es visualisiert einen vollst채ndigen Graphen $K_n$ auf der Basis eines Input-Sliders, der den User die Anzahl Knoten n ausw채hlen l채sst.

```python
import networkx as nx
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Widgets
hinweis_label = widgets.Label(value="Bitte Grad n f체r Kn w채hlen:")
my_integer_slider = widgets.IntSlider(
    value=3,          # innerhalb [min, max] starten!
    min=3,
    max=10,
    step=1,
    description="n:",
    style={"description_width": "initial"},
    layout=widgets.Layout(width="400px")
)

def draw_complete_graph(n: int):
    clear_output(wait=True)   # alten Output weg
    display(ui)               # Widgets wieder anzeigen
    G = nx.complete_graph(n)  # n wirklich benutzen
    pos = nx.spring_layout(G, seed=3)
    plt.figure()
    nx.draw(G, pos=pos, with_labels=True)
    plt.show()

def button_handler(b):
    n = my_integer_slider.value
    draw_complete_graph(n)

button = widgets.Button(description="Zeichne Graph Kn", button_style="success")

button.on_click(button_handler)

ui = widgets.VBox([hinweis_label, my_integer_slider, button])
display(ui)
```

## Graphen auf Vollst채ndigkeit pr체fen

Folgendes Skript erstellt zwei Graphen, pr체ft sie auf Vollst채ndigkeit und gibt das Ergebnis aus. 

```python
import networkx as nx
from GT_COMPLETE_CHECK import *

# Example: Create a complete graph with 4 nodes
first_graph = nx.complete_graph(4)

# Example: Create a graph with 4 nodes that is not a complete graph
second_graph = nx.Graph()
second_graph.add_nodes_from([1, 2, 3, 4])
second_graph.add_edges_from([(1, 2), (2, 3), (3, 4)])

pos = nx.spring_layout(first_graph, seed=3)
nx.draw(first_graph, pos=pos, with_labels=True)

pos = nx.spring_layout(second_graph, seed=3)
nx.draw(second_graph, pos=pos, with_labels=True, node_color="green", edge_color="green")

print("Is the first graph (blue) a complete graph?", is_complete_graph(first_graph))
print("Is the second graph (green) a complete graph?", is_complete_graph(second_graph))
```

## Nearest Neighbor Heuristik

Die Nearest Neighbor Heuristik dient dazu, L철sungen f체r das TSP Problem zu entwickeln. Sie ist eine der simpelsten Heuristiken und liefert - je nach Startpunkt - ziemlich schlechte L철sungen. 

Nachfolgendes Skript visualisiert die Funktionsweise der Nearest Neighbor Heuristik. Dazu wird eine Liste von Punkten (Knoten) im Koordinatensystem erstellt und ein Startknoten gew채hlt (z.B. Knoten 0, d.h. der erste in der Liste). Dann wird die Neighbor Heuristik angestossen und die resultierende Tour visualisiert.

```python
from GT_NEAREST_NEIGHBOR import *

# Example list of points [(x1, y1), (x2, y2), ...]
points = [(0, 0), (1, 3), (4, 6), (7, 2), (9, 5)]

# Specify the start point
start_point = 0

# Calculate the nearest neighbor tour
tour = nearest_neighbor(points, start_point)

# visualize tour
visualize_tour(tour)
```

Dummerweise liegen die Punkte etwas ung체nstig, so dass bei der Wahl von Punkt 0(0/0) als Startpunkt eine schlechte Route gefunden wird. Passen wir das Skript an, um zu sehen, was heraus kommt, wenn wir Punkt 3(4/6) als Startpunkt verwenden:

```python
# Specify the start point
start_point = 3

# Calculate the nearest neighbor tour
tour = nearest_neighbor(points, start_point)

# visualize tour
visualize_tour(tour)
```

Oh! Wir haben eine viel bessere L철sung gefunden (sogar die optimale L철sung)!

Fazit: Die Qualit채t der L철sung der NN-Konstruktionsheuristik ist abh채ngig von der Wahl des Startpunktes.
